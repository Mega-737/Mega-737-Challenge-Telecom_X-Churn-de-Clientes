# üíª1.	Extra√ß√£o dos Dados

1. Importa√ß√£o das biliotecas
2. Importa√ß√£o manual do banco de dados em Json
- Duplicado o banco de dados para gerenciar melhor as modifica√ß√µes realizadas para realizar as an√°lises.
3. Configura√ß√£o visual

```python
# Etapa 1. Importa√ß√£o de bibliotecas, configura√ß√µes visuais e carregamento dos dados JSON

# Bibliotecas principais
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tabulate import tabulate
import json
import warnings
from google.colab import files

# Configura√ß√µes de exibi√ß√£o
warnings.filterwarnings("ignore")
pd.set_option('display.max_columns', None)

# Estilo visual padr√£o para gr√°ficos
plt.rcParams['figure.figsize'] = (8, 6)
plt.rcParams['figure.dpi'] = 80
plt.rcParams['axes.edgecolor'] = 'gray'
plt.rcParams['axes.linewidth'] = 0.8
sns.set_style("whitegrid")

# Upload manual do arquivo JSON
print("Fa√ßa o upload do arquivo 'TelecomX_Data.json.json'")
uploaded = files.upload()
nome_arquivo = list(uploaded.keys())[0]

# Leitura do JSON
with open(nome_arquivo, 'r') as f:
    dados_json = json.load(f)

# Convers√£o para DataFrame
df_original = pd.json_normalize(dados_json)
df_trabalho = df_original.copy()

# Informa√ß√µes iniciais
print("‚úÖ DataFrame original carregado com sucesso!")
print("Dimens√£o do DataFrame original:", df_original.shape)
print("Dimens√£o do DataFrame de trabalho:", df_trabalho.shape)

# Visualiza√ß√£o inicial
df_original.head()
```

# ‚öôÔ∏è 2. Transforma√ß√£o

`Etapa 1`
1. Reconhecimento do Dataset: Estrutura e Colunas
2. Verificar total de colunas, linhas e nomes ‚Äî com tabulate
3. Padronizar todas as colunas para MAI√öSCULAS

```python
# Etapa 1.1. Estrutura do Dataset: Linhas, Colunas e Nomes

from tabulate import tabulate

# Total de linhas e colunas
total_linhas, total_colunas = df_trabalho.shape

# Lista com os nomes das colunas
nomes_colunas = df_trabalho.columns.tolist()

# Construindo tabela de resumo
resumo_estrutura = [
    ["Total de Linhas", total_linhas],
    ["Total de Colunas", total_colunas]
]

# Exibindo a estrutura geral
print("üìä Estrutura do DataFrame:\n")
print(tabulate(resumo_estrutura, headers=["Descri√ß√£o", "Valor"], tablefmt="grid"))

# Exibindo os nomes das colunas
print("\nüóÇÔ∏è Nomes das Colunas:\n")
print(tabulate([[col] for col in nomes_colunas], headers=["Nome da Coluna"], tablefmt="grid"))
```

```python
# Etapa 1.2. Padroniza√ß√£o dos nomes das colunas em letras MAI√öSCULAS

# Aplicando a transforma√ß√£o para mai√∫sculas nas colunas
df_trabalho.columns = df_trabalho.columns.str.upper()
df_original.columns = df_original.columns.str.upper()

# Confirma√ß√£o visual com as 5 primeiras colunas renomeadas
print("‚úÖ Nomes das colunas padronizados em MAI√öSCULAS!\n")
print("üîé Visualizando as 5 primeiras colunas ap√≥s a padroniza√ß√£o:\n")
print(tabulate([[col] for col in df_trabalho.columns[:5]], headers=["Nome da Coluna"], tablefmt="grid"))
```

`Etapa 2`
1. Tratamento de Inconsist√™ncias
2. Verificar valores ausentes
3. Verificar valores duplicados
4. Verificar formata√ß√µes e realizar ajustes
5. Garantir que todos os valores monet√°rios estejam no padr√£o brasileiro: R$
6. Garantir que colunas num√©ricas estejam como inteiros

```python
# Etapa 2.1. Verifica√ß√£o de inconsist√™ncias no banco de dados

# 1. Verifica√ß√£o de valores ausentes
valores_ausentes = df_trabalho.isnull().sum()
valores_ausentes = valores_ausentes[valores_ausentes > 0].sort_values(ascending=False)

# 2. Verifica√ß√£o de registros duplicados
duplicados = df_trabalho.duplicated().sum()

# 3. Tabela resumo das inconsist√™ncias
resumo_inconsistencias = [
    ["Total de Valores Ausentes", int(valores_ausentes.sum())],
    ["Total de Registros Duplicados", int(duplicados)]
]

# Exibi√ß√£o dos dados gerais
print("üîç Verifica√ß√£o Geral de Inconsist√™ncias:\n")
print(tabulate(resumo_inconsistencias, headers=["Tipo de Inconsist√™ncia", "Quantidade"], tablefmt="grid"))

# Exibi√ß√£o detalhada de valores ausentes por coluna (se houver)
if not valores_ausentes.empty:
    print("\nüìå Colunas com Valores Ausentes:\n")
    print(tabulate(valores_ausentes.reset_index().values,
                   headers=["Coluna", "Total de Ausentes"], tablefmt="grid"))
else:
    print("\n‚úÖ N√£o foram encontrados valores ausentes no DataFrame.")
```

```python
# Etapa 2.2. Padroniza√ß√£o de valores monet√°rios e num√©ricos

# 1. Fun√ß√£o auxiliar para formatar valores monet√°rios no estilo brasileiro (exibi√ß√£o)
def formatar_moeda(valor):
    if pd.isnull(valor):
        return "-"
    return f"R$ {valor:,.2f}".replace(",", "v").replace(".", ",").replace("v", ".")

# 2. Identificando colunas num√©ricas para convers√£o em inteiros (exceto as monet√°rias)
colunas_float = df_trabalho.select_dtypes(include=['float64']).columns

# 3. Aplicar convers√£o para inteiros nas colunas n√£o monet√°rias
for coluna in colunas_float:
    if not "CHARGES" in coluna:  # Evita aplicar nas colunas de valores monet√°rios
        df_trabalho[coluna] = df_trabalho[coluna].fillna(0).astype(int)

# 4. Exibir amostra de colunas num√©ricas convertidas
amostra_numericas = df_trabalho[colunas_float[:5]].head()

print("‚úÖ Formata√ß√£o conclu√≠da:")
print("‚úîÔ∏è Colunas num√©ricas transformadas em inteiros (exceto valores monet√°rios)")
print("\nüîé Amostra das colunas num√©ricas ap√≥s convers√£o:\n")
print(tabulate(amostra_numericas, headers="keys", tablefmt="grid", showindex=False))
```

```python
# Etapa 2.2.1. Exibir valores monet√°rios com formata√ß√£o R$ Apenas visualiza√ß√£o

# Fun√ß√£o aprimorada para formatar valores monet√°rios
def formatar_moeda(valor):
    try:
        valor = float(valor)
        return f"R$ {valor:,.2f}".replace(",", "v").replace(".", ",").replace("v", ".")
    except (ValueError, TypeError):
        return "-"

# Lista de colunas monet√°rias (baseada no nome)
colunas_monetarias = [col for col in df_trabalho.columns if "CHARGES" in col]

# Sele√ß√£o e c√≥pia para exibi√ß√£o
tabela_formatada = df_trabalho[colunas_monetarias].head().copy()
tabela_formatada = tabela_formatada.applymap(formatar_moeda)

# Exibi√ß√£o
print("üí∞ Visualiza√ß√£o das colunas monet√°rias com valores formatados (R$):\n")
print(tabulate(tabela_formatada, headers="keys", tablefmt="grid", showindex=False))
```

`Etapa 3`

**Convers√µes e C√°lculos**
1. Converter YES/NO para 1/0
2. Identificar e converter outras colunas eleg√≠veis para bin√°rio:mostrar quais
3. Somar ACCOUNT.CHARGES.TOTAL e nomear de Faturamento Total Geral
4. Criar categoria Cliente Individual e Cliente Familiar
5. Criar nova coluna ACCOUNT.DAILY.ACCOUNT com c√°lculo mensal / 30

```python
# Bloco 2.3.1 ‚Äî Convers√£o de colunas com "YES"/"NO" em bin√°rio

# Check if df_trabalho exists, if not, recreate it from df_original
if 'df_trabalho' not in globals() or df_trabalho is None:
    print("‚ö†Ô∏è df_trabalho not found. Recreating from df_original.")
    if 'df_original' in globals() and df_original is not None:
        df_trabalho = df_original.copy()
        # Ensure columns are in uppercase as done in a previous step
        df_trabalho.columns = df_trabalho.columns.str.upper()
    else:
        print("‚ùå df_original not found. Cannot proceed with conversion.")
        # You might want to add code here to reload the data if needed
        raise NameError("df_original is not defined. Please load the data.")

# 1. Identificar colunas do tipo objeto com valores YES/NO
colunas_binarias = []

for coluna in df_trabalho.select_dtypes(include='object').columns:
    valores_unicos = df_trabalho[coluna].dropna().str.upper().unique()
    if set(valores_unicos).issubset({"YES", "NO"}):
        colunas_binarias.append(coluna)

# 2. Substituir valores "YES" por 1 e "NO" por 0
for coluna in colunas_binarias:
    df_trabalho[coluna] = df_trabalho[coluna].str.upper().map({"YES": 1, "NO": 0})

# 3. Criar amostra para visualiza√ß√£o
amostra_binarias = df_trabalho[colunas_binarias].head()

# 4. Exibir colunas convertidas
print("üîÅ Convers√£o de vari√°veis 'YES'/'NO' para bin√°rio (1/0) conclu√≠da!\n")

print("üßæ Colunas convertidas:\n")
print(tabulate([[col] for col in colunas_binarias], headers=["Colunas Convertidas"], tablefmt="grid"))

print("\nüîé Amostra dos dados convertidos:\n")
print(tabulate(amostra_binarias, headers="keys", tablefmt="grid", showindex=False))
```

```python
# Etapa 3.2. C√°lculo do Faturamento Total Geral

# 1. Converter a coluna para tipo num√©rico (for√ßando erros para NaN)
df_trabalho["ACCOUNT.CHARGES.TOTAL"] = pd.to_numeric(df_trabalho["ACCOUNT.CHARGES.TOTAL"], errors="coerce")

# 2. Tratar valores ausentes com 0
df_trabalho["ACCOUNT.CHARGES.TOTAL"] = df_trabalho["ACCOUNT.CHARGES.TOTAL"].fillna(0)

# 3. C√°lculo do faturamento total
faturamento_total_geral = df_trabalho["ACCOUNT.CHARGES.TOTAL"].sum()

# 4. Exibi√ß√£o formatada
tabela_faturamento = [["Faturamento Total Geral", formatar_moeda(faturamento_total_geral)]]

print("üí∞ Faturamento Total Geral:\n")
print(tabulate(tabela_faturamento, headers=["Descri√ß√£o", "Valor (R$)"], tablefmt="grid"))
```

```python
# Etapa 3.3. Classifica√ß√£o Cliente Individual / Familiar + Faturamento

# Garantir que as colunas estejam em formato bin√°rio/integer
df_trabalho["CUSTOMER.PARTNER"] = pd.to_numeric(df_trabalho["CUSTOMER.PARTNER"], errors="coerce").fillna(0).astype(int)
df_trabalho["CUSTOMER.DEPENDENTS"] = pd.to_numeric(df_trabalho["CUSTOMER.DEPENDENTS"], errors="coerce").fillna(0).astype(int)

# Criar a nova coluna de categoria
df_trabalho["TIPO.CLIENTE"] = np.where(
    (df_trabalho["CUSTOMER.PARTNER"] == 0) & (df_trabalho["CUSTOMER.DEPENDENTS"] == 0),
    "INDIVIDUAL",
    "FAMILIAR"
)

# Agrupar por tipo de cliente e calcular totais
resumo_tipo_cliente = df_trabalho.groupby("TIPO.CLIENTE").agg(
    Total_Clientes=("TIPO.CLIENTE", "count"),
    Faturamento_Total=("ACCOUNT.CHARGES.TOTAL", "sum")
).reset_index()

# Aplicar formata√ß√£o monet√°ria para exibi√ß√£o
resumo_tipo_cliente["Faturamento_Total"] = resumo_tipo_cliente["Faturamento_Total"].apply(formatar_moeda)

# Exibir tabela
print("üè∑Ô∏è Classifica√ß√£o de Clientes por Tipo e Faturamento:\n")
print(tabulate(resumo_tipo_cliente, headers="keys", tablefmt="grid", showindex=False))
```

`Etapa 4`

**Classifica√ß√£o de Servi√ßos**
1. Criar a nova coluna GRUPO DE SERVI√áOS com os grupos: Internet, Telefone
2. Contabilizar n√∫mero de servi√ßos ativos em cada grupo e o valor que representam: ACCOUNT.CHARGES.TOTAL

```python
# Etapa 4.1. Cria√ß√£o da coluna GRUPO_DE_SERVI√áOS com classifica√ß√£o e agrupamento por tipo de servi√ßo

from tabulate import tabulate

# üîπ Definir colunas por grupo de servi√ßo
colunas_internet = [
    'INTERNET.ONLINESECURITY',
    'INTERNET.ONLINEBACKUP',
    'INTERNET.DEVICEPROTECTION',
    'INTERNET.TECHSUPPORT',
    'INTERNET.STREAMINGTV',
    'INTERNET.STREAMINGMOVIES'
]

colunas_telefone = [
    'PHONE.MULTIPLELINES'
]

# üõ†Ô∏è Criar colunas de contagem de servi√ßos "Yes" para cada grupo
df_trabalho['QTD.SERVICOS.INTERNET'] = df_trabalho[colunas_internet].apply(lambda x: (x == 'Yes').sum(), axis=1)
df_trabalho['QTD.SERVICOS.TELEFONE'] = df_trabalho[colunas_telefone].apply(lambda x: (x == 'Yes').sum(), axis=1)

# üì¶ Classificar o cliente em grupo com base nos servi√ßos ativos
def definir_grupo_servico(row):
    if row['QTD.SERVICOS.INTERNET'] > 0:
        return 'Internet'
    elif row['QTD.SERVICOS.TELEFONE'] > 0:
        return 'Telefone'
    else:
        return 'Servi√ßos Gerais'  # üõë fora da an√°lise

df_trabalho['GRUPO_DE_SERVI√áOS'] = df_trabalho.apply(definir_grupo_servico, axis=1)
```

```python
# Etapa 4.1.2. Total de Servi√ßos Ativos por Grupo e Faturamento Total

# üéØ Filtrar somente os clientes v√°lidos para an√°lise
df_validos = df_trabalho[df_trabalho['GRUPO_DE_SERVI√áOS'].isin(['Internet', 'Telefone'])].copy()

# ‚ûï Criar coluna TOTAL.SERVICOS.ATIVOS (Internet + Telefone)
df_validos['TOTAL.SERVICOS.ATIVOS'] = (
    df_validos['QTD.SERVICOS.INTERNET'] + df_validos['QTD.SERVICOS.TELEFONE']
)

# üìä Agrupar por grupo e calcular soma de servi√ßos ativos e faturamento
resumo_servicos = df_validos.groupby('GRUPO_DE_SERVI√áOS').agg({
    'TOTAL.SERVICOS.ATIVOS': 'sum',
    'ACCOUNT.CHARGES.TOTAL': 'sum'
}).reset_index()

# üìå Renomear colunas para apresenta√ß√£o
resumo_servicos.columns = [
    'Grupo de Servi√ßos',
    'Total de Servi√ßos Ativos',
    'Faturamento Total (R$)'
]

# üí∞ Formatar valores em R$
resumo_servicos['Faturamento Total (R$)'] = resumo_servicos['Faturamento Total (R$)'].apply(
    lambda x: f"R$ {x:,.2f}".replace(",", "v").replace(".", ",").replace("v", ".")
)

# üñ•Ô∏è Exibir tabela formatada
print("\nüì¶ Total de Servi√ßos Ativos e Faturamento por Grupo de Servi√ßos:\n")
print(tabulate(resumo_servicos, headers='keys', tablefmt='fancy_grid', showindex=False))
```

`Etapa 5`

**Outras An√°lises**
1. Ajustar a coluna ACCOUNT.CONTRACT
2. Levantamento de G√™nero e Perfil Et√°rio: Customer Gender e Senior

```python
# Etapa 5.1. Verificar e ajustar a coluna ACCOUNT.CONTRACT

# üîß Padroniza√ß√£o textual para garantir consist√™ncia
df_trabalho['ACCOUNT.CONTRACT'] = df_trabalho['ACCOUNT.CONTRACT'].str.strip().str.upper()

# üìä Mapeamento: contrato em meses
mapa_contrato_meses = {
    'MONTH-TO-MONTH': 1,
    'ONE YEAR': 12,
    'TWO YEAR': 24
}

# ‚ûï Nova coluna num√©rica: dura√ß√£o do contrato em meses
df_trabalho['CONTRATO.MESES'] = df_trabalho['ACCOUNT.CONTRACT'].map(mapa_contrato_meses)

# üìò Mapeamento de interpreta√ß√£o textual
mapa_interpretacao = {
    'MONTH-TO-MONTH': 'Renova√ß√£o mensal',
    'ONE YEAR': 'Contrato anual',
    'TWO YEAR': 'Contrato de dois anos'
}

# ‚ûï Nova coluna com interpreta√ß√£o em portugu√™s
df_trabalho['INTERPRETACAO.CONTRATO'] = df_trabalho['ACCOUNT.CONTRACT'].map(mapa_interpretacao)

# üìã Tabela resumo
tabela_contrato = df_trabalho[['ACCOUNT.CONTRACT', 'CONTRATO.MESES', 'INTERPRETACAO.CONTRATO']].drop_duplicates()

# üìå Exibi√ß√£o formatada
from tabulate import tabulate
print("\nüìä Tabela: Categoria de Contrato, Valor em Meses e Interpreta√ß√£o:\n")
print(tabulate(tabela_contrato, headers=['Tipo de Contrato', 'Dura√ß√£o (Meses)', 'Interpreta√ß√£o'], tablefmt='grid', showindex=False))
```

```python
# Etapa 5.2. Total por G√™nero e Perfil Et√°rio Senior

# Converter CUSTOMER.SENIORCITIZEN para "Senior" ou "N√£o Senior"
df_trabalho['PERFIL.ETARIO'] = df_trabalho['CUSTOMER.SENIORCITIZEN'].apply(
    lambda x: 'Senior' if x == 1 else 'N√£o Senior'
)

# Agrupar por g√™nero e perfil et√°rio
genero_etario = df_trabalho.groupby(['CUSTOMER.GENDER', 'PERFIL.ETARIO']).size().reset_index(name='Total')

# Renomear colunas para apresenta√ß√£o
genero_etario.columns = ['G√™nero', 'Perfil Et√°rio', 'Total']

# Exibir resultado
print("\nüë• Distribui√ß√£o por G√™nero e Perfil Et√°rio:\n")
print(tabulate(genero_etario, headers='keys', tablefmt='fancy_grid', showindex=False))
```

```python
# Etapa 6.2. Considerar somente Teste de Gr√°fico ‚Äî Faturamento por Grupo de Servi√ßos

from tabulate import tabulate

# üî¢ Dados simulados
dados_exemplo = {
    'Grupo de Servi√ßos': ['Internet', 'Telefone'],
    'Faturamento Total (R$)': [1021340.50, 529415.30]
}

# üîò Cores aplicadas do dicion√°rio da paleta
cores_barras = [
    cores_paleta['alto_risco'],  # Internet
    cores_paleta['urgente']      # Telefone
]

# üé® Gr√°fico de barras (barras finas e suaves)
fig, ax = plt.subplots()
barras = ax.bar(
    dados_exemplo['Grupo de Servi√ßos'],
    dados_exemplo['Faturamento Total (R$)'],
    color=cores_barras,
    edgecolor=cores_paleta['texto'],
    width=0.4  # üîß Barras mais finas
)

# T√≠tulo
ax.set_title('Faturamento Total por Grupo de Servi√ßos', fontsize=12, weight='bold', pad=15)

# R√≥tulos dos eixos
ax.set_xlabel('Grupo de Servi√ßos', fontsize=10)
ax.set_ylabel('Faturamento (R$)', fontsize=10)

# Valores sobre as barras
for bar in barras:
    valor = bar.get_height()
    ax.text(
        bar.get_x() + bar.get_width() / 2,
        valor + 30000,
        f"R$ {valor:,.2f}".replace(",", "v").replace(".", ",").replace("v", "."),
        ha='center',
        va='bottom',
        fontsize=8,
        color=cores_paleta['texto'],
        fontweight='bold'
    )

# Limpeza visual
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.grid(axis='y', linestyle='--', alpha=0.4)

# Renderizar
plt.tight_layout()
plt.show()
```
